# Creates directories
# Command: mkdir test{1..10}
What it does:
*   Creates directories named test1, test2, ..., up to test10 in a single command.
*   This is achieved using brace expansion, where {1..10} expands to numbers from 1 to 10
*   How to use: Type the command in your terminal and press Enter.
*   In Windows GUI: To achieve a similar result, you would manually create each folder, name it test1, test2, and so on. There's no direct equivalent to this automation in the Windows GUI.





```python
!mkdir test{1..10}
```

# ls -l deatils of the directires
Explanation of ls -l Output
When you run ls -l, you'll see a table-like format with the following columns:

File Permissions: Indicates read (r), write (w), and execute (x) permissions for the owner, group, and others.
Example: -rw-r--r--

*  - indicates a regular file
*    d indicates a directory.

*   First 3 characters: Owner permissions
*  Next 3: Group permissions.
*   Last 3: Others' permissions.





Link Count: Number of hard links to the file or directory.

Owner: Name of the user who owns the file.

Group: Name of the group that owns the file.

File Size: Size of the file in bytes.

Modification Date and Time: The last time the file or directory was modified.

File/Directory Name: The name of the file or directory.



: Shows the contents of the current folder (like viewing files in a folder in Windows Explorer).
